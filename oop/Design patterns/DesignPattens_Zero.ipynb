# Software design patterns are commonly grouped into three main categories
## Creational Design Patterns (5)

Focus: How objects are created

1. Singleton – Ensures only one instance exists

2. Factory Method – Creates objects without specifying exact class

3. Abstract Factory – Creates families of related objects

4. Builder – Builds complex objects step by step

5. Prototype – Creates new objects by copying existing ones

## Structural Design Patterns (7)

Focus: How classes and objects are composed

1. Adapter – Converts one interface into another

2. Bridge – Separates abstraction from implementation

3. Composite – Treats individual and group objects uniformly

4. Decorator – Adds behavior dynamically

5. Facade – Provides a simplified interface

6. Flyweight – Shares memory efficiently between objects

7. Proxy – Controls access to an object

## Behavioral Design Patterns (11)

Focus: How objects communicate

1. Chain of Responsibility – Passes request along a chain

2. Command – Encapsulates a request as an object

3. Interpreter – Evaluates language grammar

4. Iterator – Sequentially accesses collection elements

5. Mediator – Centralizes communication between objects

6. Memento – Saves and restores object state

7. Observer – Notifies dependent objects automatically

8. State – Changes behavior based on internal state

9. Strategy – Selects algorithm at runtime

10. Template Method – Defines skeleton of an algorithm

11. Visitor – Adds operations without modifying classes
