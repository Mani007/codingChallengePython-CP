```python
# Visitor – Separates operations from object structure.
# The Visitor Pattern lets you add new operations to objects without changing their classes.
#   Element → Accepts a visitor
#   Visitor → Performs operations on elements
```

```python
# Example: Shopping Cart 🛒
# We have:
#   Book
#   Laptop
# We want:
#   A visitor that calculates total price.
```

```python
# Visitor Interface
class Visitor:
    def visit_book(self, book):
        pass

    def visit_laptop(self, laptop):
        pass


# Concrete Visitor
class PriceCalculator(Visitor):
    def visit_book(self, book):
        return book.price

    def visit_laptop(self, laptop):
        return laptop.price


# Element Interface
class Item:
    def accept(self, visitor):
        pass


# Concrete Elements
class Book(Item):
    def __init__(self, price):
        self.price = price

    def accept(self, visitor):
        return visitor.visit_book(self)


class Laptop(Item):
    def __init__(self, price):
        self.price = price

    def accept(self, visitor):
        return visitor.visit_laptop(self)


# Usage
items = [
    Book(30),
    Laptop(1000),
    Book(20)
]

visitor = PriceCalculator()

total = 0
for item in items:
    total += item.accept(visitor)

print("Total price:", total)
```

```python
# What’s Happening?
# Each element (Book, Laptop) implements accept().
# The visitor (PriceCalculator) contains the operation logic.
# Adding a new operation (e.g., DiscountCalculator) requires creating a new visitor — without modifying the item classes.
```
