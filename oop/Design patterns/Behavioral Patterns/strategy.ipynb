```python
# Strategy – Encapsulates interchangeable algorithms.
# The Strategy Pattern lets you define a family of algorithms, put each one in a separate class, and make them interchangeable at runtime.
# Instead of using multiple if/else statements, we:
# Create separate strategy classes
# The context uses whichever strategy is assigned
```

```python
# Example: Payment System 💳
# Different payment methods:
#   Credit Card
#   PayPal
#   Cash
```

```python
# Strategy Interface
class PaymentStrategy:
    def pay(self, amount):
        pass


# Concrete Strategies
class CreditCardPayment(PaymentStrategy):
    def pay(self, amount):
        print(f"Paid {amount} using Credit Card")


class PayPalPayment(PaymentStrategy):
    def pay(self, amount):
        print(f"Paid {amount} using PayPal")


class CashPayment(PaymentStrategy):
    def pay(self, amount):
        print(f"Paid {amount} using Cash")


# Context
class ShoppingCart:
    def __init__(self, strategy: PaymentStrategy):
        self.strategy = strategy

    def set_strategy(self, strategy: PaymentStrategy):
        self.strategy = strategy

    def checkout(self, amount):
        self.strategy.pay(amount)


# Usage
cart = ShoppingCart(CreditCardPayment())
cart.checkout(100)

cart.set_strategy(PayPalPayment())
cart.checkout(200)

cart.set_strategy(CashPayment())
cart.checkout(50)
```

```python
# The ShoppingCart doesn’t care how payment is done.
# You can change behavior at runtime using set_strategy().
# No if/else chains needed.
```
