```python
# Template Method – Defines algorithm skeleton, subclasses fill steps.
# The Template Method Pattern defines the skeleton of an algorithm in a base class, but lets subclasses override specific steps without changing the overall structure.
# The base class controls the flow.
# Subclasses customize certain steps.
```

```python
# Example: Making a Beverage 
# Steps:
#   Boil water
#   Brew
#   Pour into cup
#   Add condiments
# The structure stays the same, but brewing and condiments differ.
```

```python
# Abstract Class
class Beverage:

    # Template Method (defines algorithm structure)
    def make(self):
        self.boil_water()
        self.brew()
        self.pour_in_cup()
        self.add_condiments()

    def boil_water(self):
        print("Boiling water")

    def pour_in_cup(self):
        print("Pouring into cup")

    # Steps to be implemented by subclasses
    def brew(self):
        pass

    def add_condiments(self):
        pass


# Concrete Class 1
class Tea(Beverage):
    def brew(self):
        print("Steeping the tea")

    def add_condiments(self):
        print("Adding lemon")


# Concrete Class 2
class Coffee(Beverage):
    def brew(self):
        print("Brewing the coffee")

    def add_condiments(self):
        print("Adding sugar and milk")


# Usage
tea = Tea()
coffee = Coffee()

print("Making Tea:")
tea.make()

print("\nMaking Coffee:")
coffee.make()
```

```python
# What’s Happening?
# make() is the template method — it defines the fixed sequence.
# Subclasses override only specific steps.
# The overall process cannot be changed by subclasses.
```
