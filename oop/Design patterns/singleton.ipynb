```python
# Singleton ensures that a class has only one instance and provides a global access point to it.
class Singleton:
    _instance = None  # class variable to store single instance

    def __new__(cls):
        if cls._instance is None:
            print("Creating new instance")
            cls._instance = super().__new__(cls)
        return cls._instance
```

```python
obj1 = Singleton()
obj2 = Singleton()

print(obj1 is obj2)  # True
```

```python
# __new__() controls object creation.
# First time → creates object.
# Next time → returns same object.
# Both variables point to the same instance.
```
