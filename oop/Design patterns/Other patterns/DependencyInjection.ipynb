```python
# Instead of a class creating its dependencies, they are provided (injected) from outside.
# This makes code:
#   Easier to test
#   Loosely coupled
#   More flexible
```

```python
# Without Dependency Injection (Tightly Coupled)
# Problem: UserService is forced to use MySQLDatabase.
class MySQLDatabase:
    def connect(self):
        print("Connected to MySQL")


class UserService:
    def __init__(self):
        self.db = MySQLDatabase()  # tightly coupled

    def process(self):
        self.db.connect()
        print("Processing user data")


service = UserService()
service.process()
```

```python
# With Dependency Injection (Constructor Injection)
# Dependency Interface
class Database:
    def connect(self):
        pass


# Concrete Implementations
class MySQLDatabase(Database):
    def connect(self):
        print("Connected to MySQL")


class PostgreSQLDatabase(Database):
    def connect(self):
        print("Connected to PostgreSQL")


# Dependent Class
class UserService:
    def __init__(self, database: Database):
        self.db = database  # injected dependency

    def process(self):
        self.db.connect()
        print("Processing user data")


# Injecting dependency
mysql_db = MySQLDatabase()
postgres_db = PostgreSQLDatabase()

service1 = UserService(mysql_db)
service1.process()

print("---- Switching database ----")

service2 = UserService(postgres_db)
service2.process()
```

```python
# What’s Happening?
# UserService does not create the database.
# The database is injected from outside.
# You can switch implementations easily.
# You can inject a MockDatabase for testing.
```
