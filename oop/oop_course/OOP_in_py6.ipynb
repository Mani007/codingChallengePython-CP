```python
# 1. Simple inheritance
class Base:
    def __init__(self,id):
        self.id = id
    def printId(self):
        print(f"From the base class with {self.id}")
class Derived(Base):
    def __init__(self, id):
        super().__init__(id)
    def printId(self):
        return super().printId()  # calling the base class method
d1 = Derived(10)
d1.printId()
```
