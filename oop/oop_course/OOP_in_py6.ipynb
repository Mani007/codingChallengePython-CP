```python
# Types of Inheritance
# 1. Simple Inheritance
# 2. HIERARCHY inheritance 
# 3. Multilevel inheritance
# 4. Multiple inheritance
# 5. Hybrid inheritance - Diamond pattern
```

```python
# 1. Simple inheritance
class Base:
    def __init__(self,id):
        self.id = id
    def printId(self):
        print(f"From the base class with {self.id}")
class Derived(Base):
    def __init__(self, id):
        super().__init__(id)
    def printId(self):
        return super().printId()  # calling the base class method
d1 = Derived(10)
d1.printId()
```

```python
# Hierarchy Inheritance
class Grandfather:
    def __init__(self,id):
        self.id = id
    def printId(self):
        print(f"From the Grandfather with id {self.id}")
class Parent(Grandfather):
    def __init__(self, id):
        super().__init__(id)
    def printId(self):
        return super().printId()
    
class Child(Parent):
    def __init__(self, id):
        super().__init__(id)
    def printId(self):
        return super().printId()
c1 = Child(15)
c1.printId()
        
```

```python
# Multilevel Inheritance
class Grandfather:
    def __init__(self,id):
        self.id = id
    def printId(self):
        print(f"From the Grandfather with id {self.id}")
class Parent(Grandfather):
    def __init__(self, id):
        super().__init__(id)
    def printId(self):
        return super().printId()
    
class Child(Parent):
    def __init__(self, id):
        super().__init__(id)
    def printId(self):
        return super().printId()
c1 = Child(15)
c1.printId()
        
```

```python
# Multiple Inheritance
class Grandfather:
    def __init__(self,id):
        self.id = id
    def printId(self):
        print(f"From the Grandfather with id {self.id}")
class A(Grandfather):
    def __init__(self, id):
        super().__init__(id)
    def printId(self):
        print("From A")
        #return super().printId()
class B(Grandfather):
    def __init__(self, id):
        super().__init__(id)
    def printId(self):
        print("From B")
        return super().printId()
    
class Child(A,B):
    def __init__(self, id):
        super().__init__(id)
    def printId(self):
        return super().printId()
c1 = Child(15)
c1.printId()
        
```
