```python
# Inheritance
class Auth:
    def __init__(self,name,id):
        self.name = name 
        self.id = id
    def login(self):
        print("You are logged in !!")
    def __logout(self): # maing logout method protected using single underscore and only child can use it not iny instance of Auth class
        print("Logout successfully, Thanks !!")
```

```python
class Student(Auth): # Inheritance, Student inherited from Auth
    def __init__(self,marks,rollno):
        #super().__init__(name, id)
        self.marks = marks
        self.rollno = rollno
    def login(self): # This login method will override the Auth login method
        print("STUDENT login")
    def play(self):
        print("playing!!!")
```

```python
s1 = Student("abc",11)
# We inherit, 1. Non-private variable and methods 2. constructors and 3. Magic methods
# Parent cannot inherit from child
```

```python
s1.login()  # Login method of student is called
s1.play() # we cannot use protected directly here
```
