```python
# Inheritance
class Auth:
    def login(self):
        print("You are logged in !!")
    def _logout(self): # maing logout method protected using single underscore and only child can use it not iny instance of Auth class
        print("Logout successfully, Thanks !!")
```

```python
class Student(Auth): # Inheritance, Student inherited from Auth
    pass
```

```python
s1 = Student()
# We inherit, 1. Non-private variable and methods 2. constructors and 3. Magic methods
# Parent cannot inherit from child
```

```python
s1.login()
s1.logout() # we cannot use protected directly here
```
