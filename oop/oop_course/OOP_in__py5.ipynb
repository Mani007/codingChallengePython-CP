```python
# Inheritance
class Auth:
    def __init__(self,name,id):
        self.name = name 
        self.id = id
    def login(self):
        print("You are logged in !!")
    def __logout(self): # maing logout method protected using single underscore and only child can use it not iny instance of Auth class
        print("Logout successfully, Thanks !!")
```

```python
class Student(Auth): # Inheritance, Student inherited from Auth
    def __init__(self,marks,rollno,name,id):
        # Super cannot ne used outside the class
        super().__init__(name, id) # Super is method to access parent from child
        self.marks = marks
        self.rollno = rollno
    def login(self): # This login method will override the Auth login method
        print("STUDENT login")
    def play(self):
        newname = super().name # super cannot be used to access the parent attribute 
        print("playing!!!")
        print(newname)
```

```python
s1 = Student(95,1,"abc",11)
print(s1.marks)
s1.name
# We inherit, 1. Non-private variable and methods 2. constructors and 3. Magic methods
# Parent cannot inherit from child
```

```python
s1.play()
```

```python
s1.login()  # Login method of student is called
s1.play() # we cannot use protected directly here
```

```python
class Seller:
    def __init__(self):
        print("Seller constructor")
        self.name = "abc"
        self._website = "www.aaa.com" # protected member can be accessed in inherited class
        self.__code = "XYZ" # private member - cannot be accessed outside the class
    def login(self):
        print("Seller Login")
    def logout(self):
        print("Seller logout")
```

```python
class Buyer(Seller): # inheriting seller
    def __init__(self):
        print("Buyer Constructor")
        super().__init__()
        self.name = "PQR"
    def login(self):
        super().logout()
    def access(self):
        print(self.name)
        print(self._website)
        print(self.__code) # cannot access this its private
    
```
