```python
# Polymorphism in Python
# 1. Method overloading
# 2. Method overriding
# 3. Operator overloading
```

```python
# method overloading - same method name but different argument (Not by default supported in python)
# This method overloading problem is due to python being interpreted language, so in complied language c++/Java this problem will not occur
def sum2(a,b):
    print(a+b)
```

```python
sum2(2,3)
```

```python
def sum3(a,b,c):
    print(a+b+c)
```

```python
# fixing method overloading
def sumBetter(a,b,c=0):
    if c==0:
        sum2(a,b)
    else:
        sum3(a,b,c)
```

```python
sumBetter(1,2,3)
```

```python
sumBetter(3,4)
```

```python
# Better way method overloading
def sum(*args):
    count = 0
    for i in args:
        count +=i
    print(count)
```

```python
sum(1,2,3)
```

```python
sum(1,2,3,4)
```
