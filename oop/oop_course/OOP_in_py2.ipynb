```python
# OOP creating a complex class
'''
Creating an complexNumber class that perform operations such as add,sub,mul,div,conjugate etc. 
'''
class ComplexNumber:
    def __init__(self,real = 0,imag = 0): # default parameter in constructor
        self.real = real
        self.imag = imag
```

```python
cn = ComplexNumber()
cn.real,cn.imag
```

```python
cn1 = ComplexNumber(2,5)
cn1.real,cn1.imag
```
