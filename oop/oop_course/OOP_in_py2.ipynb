```python
# OOP creating a complex class
'''
Creating an complexNumber class that perform operations such as add,sub,mul,div,conjugate etc. 
'''
class ComplexNumber:
    def __init__(self,real = 0.0,imag = 0.0): # default parameter in constructor
        self.real = real
        self.imag = imag
    def __str__(self): # string representation is defined here
        if self.real == 0:
            s = f"{self.imag}i"
        elif self.imag<0:
            s = f"{self.real}{self.imag}i"
        else:
            s = f"{self.real}+{self.imag}i"
        return s  
    def __repr__(self):
        if self.real == 0:
            s = f"{self.imag}i"
        elif self.imag<0:
            s = f"{self.real}{self.imag}i"
        else:
            s = f"{self.real}+{self.imag}i"
        return s  
    def __add__(self,other): # it is a binary operator for exaple 2+3+4 =  (2+3)+4
        resultReal = 0
        resultImag = 0
        resultReal = self.real + other.real      
        resultImag = self.imag + other.imag   
        ans = ComplexNumber(resultReal,resultImag)
        return ans   
    def conjugate(self):
        '''
        this method return conjugate of a complex number. 
        example: 2+3j conjugate is 2-3j
        make sure that you do not change the original complex number 
        We need to override using magic/dunder methods to have this method completely working. 
        __str__ is the method that calls the print method, currently this method is operating on default. We need to override this.
        '''
        self.imag = self.imag*(-1)
        if self.real == 0:
            s = f"{self.imag}i"
        elif self.imag<0:
            s = f"{self.real}{self.imag}i"
        else:
            s = f"{self.real}+{self.imag}i"
        return s 
```

```python
cn = ComplexNumber(0,-2)
cn.conjugate()
print(cn)
```
