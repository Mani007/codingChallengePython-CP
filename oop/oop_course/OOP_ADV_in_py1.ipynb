```python
# This is the concept of first class function, closure, decorator, class-making a.k.a meta-classes
# we can use any form of class and function nesting for multiple layers
def make_class(x):
    class Base:
        def __init__(self,name):
            self.name = name
        def printId(self):
            print(f"The current id is {x}")
    return Base    # we are returning class reference here not the instance
    #return Base() #Please note that we are not returning the instance of the class            
```

```python
some_base = make_class(10)
```

```python
some_base("Abc") # We need to store this to some variable for class instance
```

```python
some_base.printId() # we cannot directly do this as this is just the class reference and constructor not initiated
```

```python
store_print = some_base.printId
```

```python
base_instance = some_base("ABC") # initiated the constructor of the Base class and store it to instance
```

```python
base_instance.printId()
```

```python
base_instance.name
```

```python
# returning function from function
def make_function(x):
    if x == 1:
        def rf():
            print("Return Value is 1")
    else:
        def rf():
            print("Return Value is NOT 1")
    return rf 
       
```

```python
func_instance = make_function(1)
func_instance()  # calling the function rf from function instance
```

```python
func_instance = make_function(11)
func_instance()  # calling the function rf from function instance
```
