```python
# ENCAPSULATION in Python
# Access Modifiers in Python
class Student:
    # class variable here
    __count = 0
    def __init__(self,name,marks):
        # instance variable
        self.name = name
        self.__marks = marks  # We have made this variable private and we need getter and setter to handle it. 
        self.numberOfStudent = Student.__count + 1 # storing the class variable as instance variable
        # updating the class variable
        Student.__count += 1
    def __auth(self,passCode): # Private authentication function used only internally in class
        if passCode == 0000:
            return True
        else:
            return False
    def getMarks(self):
        print(self.__marks)
    def setMarks(self,newMarks,passCode): # Adding additional security to the private member 
        if self.__auth(passCode) == True:
            self.__marks = newMarks
            print("update successful")
        else:
            print("Cannot perform, you dont have access")
s1 = Student("abc",45) 
s2 = Student("pqr",55) 
s3 = Student("xyz",95)
#print(Student.__count) # We cannot do this its private now
#print(s1.count) 
print(s2.numberOfStudent) 
        
```
