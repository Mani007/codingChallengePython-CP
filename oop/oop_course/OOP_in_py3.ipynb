```python
# ENCAPSULATION in Python
# Access Modifiers in Python
class Student:
    # class variable here
    count = 0
    def __init__(self,name,marks):
        # instance variable
        self.name = name
        self.__marks = marks  # We have made this variable private and we need getter and setter to handle it. 
        self.numberOfStudent = Student.count + 1 # storing the class variable as instance variable
        # updating the class variable
        Student.count += 1
    def getMarks(self):
        print(self.__marks)
s1 = Student("abc",45) 
s2 = Student("pqr",55) 
s3 = Student("xyz",95)
print(Student.count) 
print(s1.count) 
print(s2.numberOfStudent) 
        
```
