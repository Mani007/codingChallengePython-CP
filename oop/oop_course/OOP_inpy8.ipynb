```python
# Abstraction - use it without getting in internal details
# python has Abstract classes and abstract methods to support abstraction
# We have to ensure two things
# 1. import abc class
# 2. should have abstract method in abstract class 
```

```python
from abc import ABC,abstractmethod
class User(ABC): # User class is also known as abstract base class, you can only inherit it and implement its method according to your need
    @abstractmethod # abstract classes must have at least one abstractmethod 
    def login(self): # responsibility of child class to implement the login method as they want
        pass
```
