```python
# Understand static variable in python
class Student:
    __schoolName = "KV" # class variable or static property
    def __init__(self,name):
        self.__name = name
    @staticmethod
    def getSchool():# We do not need self here as its static/class attribute
        return Student.__schoolName
    @staticmethod
    def setSchool(newSchool):# We do not need self here as its static/class attribute
        Student.__schoolName = newSchool
    
    def getInstanceSchool(self):# We do not need self here as its static/class attribute
        return Student.__schoolName
s1 = Student("ABC")
```

```python
s1.setSchool("Myschool")
```

```python
s1.getSchool() # Now this will work without self as its a static method 
```

```python
s1.getInstanceSchool()  # Removed @staticmethod and used self
```
