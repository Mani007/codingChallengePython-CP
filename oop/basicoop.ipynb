```python
class Base:
    '''
    Docstring for Base: Here is the base class
    example: 
    b = Base(data)
    print(b.data)
    output:
    data
    '''
    def __init__(self,data):
        '''
        Constructor for Base class
        args:
            data: any object
        return:
            Nothing, just use to set the object variable
        '''
        self.data = data
    
    def add_two(self):
        '''
        Add 2 to the data
        Use: 
        It adds two to the class instance variable data. 
        args: 
            Nothing
        return:
            Nothing
        '''
        self.data += 2
```

[Python Docstring](https://www.geeksforgeeks.org/python/python-docstrings/)

```python
bb = Base(25)
print(bb.data)
```

```python
help(bb)
```

```python
bb.add_two()
print(bb.data)
```
