```python
class Base:
    '''
    Docstring for Base: Here is the base class
    example: 
    b = Base(data)
    print(b.data)
    output:
    data
    '''
    def __init__(self,data):
        self.data = data
    
```

```python
bb = Base(25)
print(bb.data)
```
