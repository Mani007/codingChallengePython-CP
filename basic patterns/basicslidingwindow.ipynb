```python
# fixed size sliding window
st = 'abcdeabcc'
# fix size of 3 sliding
left = 0
right = 0
l = len(st)
triplet = []
while  right<l:
    print(st[left])
    left += 1
    right += 1
```

```python
# fixed size sliding window
st = 'abcdeabcc'
# fix size of 3 sliding
l = len(st)
triplet = []
while  right<=l:
    left = 0
    right = left+3
    for s in range(right-left):
        triplet.append(st[left])
    left += 1
    right += 1
print(triplet)
```
