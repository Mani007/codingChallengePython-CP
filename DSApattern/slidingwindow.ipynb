```python
# Find longest common subarray from a string
# string operations
st = "abcabcbb"
print(st[1])
for s in st:  # without typecasting as string is iterable
    print(s)
```

```python
def longsubarray(s:str)->int:
    left = 0
    longest = 0
    sett = set()
    n = len(s)
    #windowsize = 0
    for right in range(n):
        while s[right] in sett:
            sett.remove(s[left])
            left +=1 # moving left pointer
        #right += 1
        windowsize = (right - left) + 1
        longest = max(longest,windowsize)
        sett.add(s[right])
    return longest
#print(int(longsubarray(st)))
res = longsubarray(st)
print(res)
```
