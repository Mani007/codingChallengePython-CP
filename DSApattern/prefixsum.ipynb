# Prefix sum

```python
import random
li = random.sample(range(1, 101), 5)
print(li)
```

```python
# Generate Random number list
# random_list = [x for x in range(50)] 
random_list = random.sample(range(1, 101), 10)
print(random_list)
```

```python
# Implementation of prefix sum
sum = [0]*(len(random_list)+1) # Prefix sum array intialized to 0 with one extra index
#print(sum)
for i in range(len(random_list)):
    sum[i+1] = sum[i]+random_list[i]   # Rule for creating prefix sum array/list
print(sum)
```

```python
# Solving problem for prefix sum 
# from index 6 to 9
res = sum[9] - sum[6]
print(res)
```
