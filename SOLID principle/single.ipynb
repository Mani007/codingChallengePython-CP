```python
# Violation of single responsibility principle
# Doing too much in single class
class Bank:
    def __init__(self,acc,amount,actype, bal):
        self.acc = acc
        self.amount = amount
        self.type = actype
        self.bal = bal
    def deposit(self,damount):
        self.amount  += damount
    def withdraw(self,wamount):
        self.amount  -= wamount

    
```

```python
# Single responsibility principle
class Account:
    def __init__(self, acc):
        self.__acc = acc
    # Getter
    @property
    def acc(self):
        return self.__acc

    # Setter
    @acc.setter
    def name(self, value):
        self.__acc = value
        
class Deposit(Account):
    pass
class Withdraw(Account):
    pass
```
